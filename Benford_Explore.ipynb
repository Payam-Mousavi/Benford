```python
import numpy as np
import random
import matplotlib.pyplot as plt

random.seed(123)
```

```python
## Generate 1000 Random integers in range [1,max_val]
N = 10000 # Number of samples
max_val = 10000 # Maximum integer value
rand_ints = [random.randrange(1, max_val) for i in range(N)]

# Plotting:
plt.xlabel('Bins')
plt.ylabel('Counts')
plt.title('Uniformly distributed randomly generated numbers')
plt.hist(rand_ints, bins='auto')
plt.show()
```

```python
## Get the first digits only:
first_digits = [int(str(rand_ints[i])[0]) for i in range(N)]

# Plotting:
plt.xlabel('Bins')
plt.ylabel('Counts')
plt.title('Distribution of first digits')
plt.hist(first_digits, bins='auto')
plt.show()
```

## Try with Gaussian distributions:

```python
rand_ints_gaussian = [round(np.random.normal(0, 100)) for i in range(N)]
# Taking the absolute value of the negative samples
first_digits_gaussian_v1 = [int(str(abs(rand_ints_gaussian[i]))[0]) for i in range(N)]

# Not counting the negative samples:
first_digits_gaussian_v2 = [int(str(rand_ints_gaussian[i])[0]) for i in range(N) if rand_ints_gaussian[i]>=0]
    


# Plotting:

figure, axes = plt.subplots(nrows=1, ncols=2)
# figure.tight_layout()
plt.subplot(2,1,1)
plt.xlabel('Bins')
plt.ylabel('Counts')
plt.title('Distribution of first digits (Gaussian) - Absolute Value taken')
plt.hist(first_digits_gaussian_v1, bins='auto')
plt.show()

plt.subplot(2,1,2)
plt.xlabel('Bins')
plt.ylabel('Counts')
plt.title('Distribution of first digits (Gaussian) - Negatives removed')
plt.hist(first_digits_gaussian_v2, bins='auto')
plt.show()
```
